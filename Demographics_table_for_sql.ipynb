```python
import pandas as pd
```

```python
Path = 'income_per_code.csv'
```

```python
df_income = pd.read_csv(Path)
```

```python
df_income
```

```python
df_income['ZIP Code'] = df_income['ZIP Code'].astype('string')
```

```python
Path2 = 'pop_density_per_zip.csv'
```

```python
df_popul = pd.read_csv(Path2)
```

```python
df_popul
```

```python
df_popul.rename(columns={'Zipcode' : 'ZIP Code'}, inplace = True)
```

```python
df_popul.info()
```

```python
df_popul['ZIP Code'] = df_popul['ZIP Code'].astype('string')
```

```python
df_popul.info()
```

```python
columns_to_drop = [
    'Land Area sq meters',
    'Water Area sq meters',
    'Population',
    'Population Density per sq mile'
]
df_popul= df_popul.drop(columns=columns_to_drop)
```

```python
df_popul
```

```python
Path3 = 'poverty_rate_per_zip.csv'
```

```python
df_poverty = pd.read_csv(Path3)
```

```python
df_poverty
```

```python
df_poverty.rename(columns={'Zipcode' : 'ZIP Code'}, inplace = True)
```

```python
df_poverty.info()
```

```python
df_poverty['ZIP Code'] = df_poverty['ZIP Code'].astype('string')
```

```python
df_poverty.info()
```

```python
df_inc_pop = pd.merge(df_income, df_popul, on='ZIP Code', how='outer')

df_inc_pop
```

```python
df_inc_pop_pov = pd.merge(df_inc_pop, df_poverty, on='ZIP Code', how='outer')
df_inc_pop_pov
```

```python
df_inc_pop_pov = df_inc_pop_pov.drop('Population', axis=1)
```

```python
df_inc_pop_pov.info()
```

```python
df_inc_pop_pov['Longitude'] = df_inc_pop_pov['Longitude'].replace('−', '-', regex=True)
df_inc_pop_pov['Longitude'] = df_inc_pop_pov['Longitude'].astype('float')

# Reset the index to create a new column with auto-incrementing values
df_inc_pop_pov.reset_index(inplace=True)

# Rename the new column to 'ID'
df_inc_pop_pov.rename(columns={'index': 'Pop_ID'}, inplace=True)

df_inc_pop_pov
```

```python
df_inc_pop_pov.info()
```

```python
Path_Doctors = "Final_doctors.csv"
doctors_df = pd.read_csv(Path_Doctors)
doctors_df.rename(columns ={'Zip Code': 'ZIP Code'}, inplace = True)

doctors_df.drop('Primary Practice', axis=1, inplace=True)

doctors_df['ZIP Code'] = doctors_df['ZIP Code'].astype('string')

doctors_df.head()
```

```python
Demographics_df = pd.merge(df_inc_pop_pov, doctors_df, on='ZIP Code', how='outer')

Demographics_df.head(10)
```

```python

Demographics_df.to_csv("Demographics_for_sql.csv", index = False)
```
